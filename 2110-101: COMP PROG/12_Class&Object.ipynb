<ins>note</ins> เราจะเรียก function ใน class ว่า method

````python
class Point:
    def __init__(self, x, y):
        self.x = x      #ตัวแปรที่ส่งเข้ามาหรือใช้ใน class จะเป็นชนิดไหนก้ได้
        self.y = y      #จะเป็น int, str, list หรือ object ก็ได้

p1 = Point(3,4)
print("p1.x:", p1.x)
print("p1.y:", p1.y)
p2 = Point(5,6)
print("p2.x:", p2.x)
print("p2.y:", p2.y)
print(p1)
````

## **การแสดงเนื้อหาของ Object**

````python
import math
class Circle:
    def __init__(self, r):
        self.r = r
        self.area = math.pi * r**2
        self.diameter = math.pi * r * 2

    def to_str(self):
        return "Circle with radius: " + str(self.r)

c = Circle(2)
print(c.to_str()) # เรียกใช้โดย object.method
````

คำสั่ง print() และ str() จะไปเรียก method \_\_str\_\_() ของ Object ซึ่งเราสามารถกำหนดได้ว่าให้คำสั่งนอก class เห็นเนื้อหาอะไรใน class ได้บ้าง

\_\_str\_\_() ควรจะเขียนให้อ่านเข้าใจง่าย

````python
import math
class Circle:
    def __init__(self, r):
        self.r = r
        self.area = math.pi * r**2
        self.diameter = math.pi * r * 2

    def __str__(self):
        return "Circle with radius: " + str(self.r)

c = Circle(2)
print(c)
print(str(c))
print(c.__str__())
````

````python
class MyNumber:
    def __init__(self, x):
        self.num = x

    def __add__(self, y):
        return MyNumber(int(str(self.num) + str(y.num)))

    def __lt__(self, y):
        return self.num < y.num

    def __str__(self):
        return "my number is: " + str(self.num)

m1 = MyNumber(2)
m2 = MyNumber(3)
print(m1+m2)
print(m1.__add__(m2))
print(m1 < m2)
print(m1 > m2)
````

### **```__str__``` vs ```__repr__```**

คำจำกัดความของ ```str()``` จาก [doc.python.org](https://docs.python.org/3/library/stdtypes.html#str) ไม่ออกสอบ (แต่รู้ไว้ก็ดีนะ)

```
Return a string version of object. If object is not provided, returns the empty string. Otherwise, the behavior
of str() depends on whether encoding or errors is given, as follows.

If neither encoding nor errors is given, str(object) returns type(object).__str__(object), which is the
“informal” or nicely printable string representation of object. For string objects, this is the string itself. If object
does not have a __str__() method, then str() falls back to returning repr(object).
```

คำจำกัดความของ ```repr()```

```
Return a string containing a printable representation of an object. For many types, this function makes an
attempt to return a string that would yield an object with the same value when passed to eval(); otherwise,
the representation is a string enclosed in angle brackets that contains the name of the type of the object
together with additional information often including the name and address of the object. A class can control
what this function returns for its instances by defining a __repr__() method.
```

แปลสั้นๆได้ว่า ```__str__()``` จะถูกเรียกโดย ```str()``` โดยเป้าหมายคืออ่านง่าย และถ้า ```__str__()``` ไม่ได้เขียนไว้ ```__repr__()``` จะถูกเรียกแทน

ส่วน ```__repr__()``` จะถูกเรียกโดย ```repr()``` โดยมีเป้าหมายเพื่อไว้ใช้สร้างตัวมันเอง ถ้าถูกเรียกด้วย `eval()`

````python
# eval คืออะไร
x = [7, 8, 9]
y = eval("[7, 8, 9]")
print(x == y)
print(x)
print(y)
print(type(y))
#print(x == repr(x))
#print(x == eval(repr(x)))
````

````python
x = (6,)
print(x) # (6)
print(x == eval(repr(x)))
````

````python
# repr vs str
import numpy as np
from numpy import array

x = np.array([[4,2,2,2],
              [1,0,0,0],
              [1,0,0,0],
              [1,1,0,0]])
print(x)
print(repr(x))
# li = list(np.dfdfd)
# "["+repr(li[1])
print(eval(repr(x)))
print(x == eval(repr(x)))
````

````python
# การเขียน method __str__ และ __repr__
class Point:
    def __init__(self, x, y):
        self.x = x      #ตัวแปรที่ส่งเข้ามาหรือใช้ใน class จะเป็นชนิดไหนก้ได้
        self.y = y      #จะเป็น int, str, list หรือ object ก็ได้
    def __str__(self):
        return f"({self.x},{self.y})"
    def __repr__(self):
        return f"Point({self.x},{self.y})"
    def __eq__(self, p):
        return self.x == p.x and self.y == p.y

p1 = Point(3, 4)
print(p1)
print(p1 == (3,4))
print(repr(p1))
p2 = eval(repr(p1))
print(p2)
print(p1 == p2)
````

## **การเขียน method ให้กับคลาส การเขียน method ให้อ็อบเจกต์**

ไม่มีใน slide (ไม่ออกสอบ)

````python
# use of a class method and static method.
from datetime import date

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Person is adult or not.
    def isAdult(self):            # instance method
        return self.age > 18

    # Person is adult or not.
    @staticmethod
    def is_Adult(age):             # static method
        return age > 18

p1 = Person('john', 21)
print(p1.isAdult())

print(Person.is_Adult(22))
````
