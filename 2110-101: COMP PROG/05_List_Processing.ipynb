## **append**

เป็นการเพิ่มข้อมูลเข้าไปท้าย list
> append(value)





```python
# basic
x = ["apple", "banana", "cherry"]
x.append(3) # append ข้อมูลคนละชนิดกับที่มีอยู่ก็ได้
print(x)
x.append(True)
print(x)
x.append("4")
print(x)
x.append("orange")
print(x)

y = []
y.append("hello")
print(y)
```

## **remove**

ใช้ในการลบค่าออกจาก list
> remove(value)

```python
# basic
x = ["apple", "banana", "cherry"]
x.remove("banana")
print(x)
```

## **insert**

เพิ่มค่าเข้าไปในตำแหน่งที่กำหนด
> insert(pos, value)

```python
# insert 20 at the beginning of the list
x = ["apple", "banana", "cherry"]
x.insert(0, 20)
print(x)
```

```python
# insert 20 at index 2 of the list
x = ["apple", "banana", "cherry"]
x.insert(2, 20)
print(x)
```

```python
# insert 20 at the end of the list (this is the same as append)
x = ["apple", "banana", "cherry"]
x.insert(len(x), 20)
print(x)
```

```python
# insert 20 beyond the end of the list
x = ["apple", "banana", "cherry"]
x.insert(8, 20)
print(x)
```

```python
# insert 20 before the beginning of the list
x = ["apple", "banana", "cherry"]
x.insert(-5, 20)
print(x)
```

## **pop**

ลบค่าจาก list ในตำแหน่งที่กำหนดไว้ และคืนค่านั้นออกมา
> remove(index)

ถ้าไม่ระบุตำแหน่ง จะลบค่าตัวสุดท้ายและคืนค่าตัวสุดท้ายออกมา

```python
x = ['apple', 'banana', 'cherry', 'orange', 'grape']
y = x.pop(1)
print(x)
print(y)
y = x.pop()
print(x)
print(y)
```

## **sort**
เรียงค่าใน list
> sort()

ถ้าไม่กำหนดค่าใดๆ จะเรียงจากน้อยไปหามาก

# **Extra: Mutable vs Immutable**

Objects of built-in type that are mutable are:
- Lists
- Sets (*)
- Dictionaries (*)
- User-Defined Classes (It purely depends upon the user to define the characteristics) (**)

Objects of built-in type that are immutable are:
- Numbers (Integer, Float)
- Booleans
- Strings
- Tuples (*)
- User-Defined Classes (It purely depends upon the user to define the characteristics) (**)

Data types with (\*) will be seen in later chapters

Data types with (\*\*) will be covered in this course but you are unlikely to clearly see how to make it mutable/immutable


## **Immutable**

In Python, if the value of an object cannot be changed over time, then it is known as immutable. Once created, the value of these objects is permanent.

คำแปล #1: เมื่อกำหนดค่าให้ตัวแปรประเภท Immutable แล้วจะเปลี่ยนแปลงไม่ได้

คำแปล #2: ค่าของตัวแปร Immutable จะเปลี่ยนได้ก็ต่อเมื่อตัวแปรนั้นอยู่ทางซ้ายของเครื่องหมาย `=` (ซึ่งคือถูกกำหนดค่าใหม่นั่นแหละ)

```python
# We will use id function, which return the address of variable, to visualize this
a = 2
print(f"id(a): {id(a)}")
b = a
print(f"**** after b = a ****\nid(a): {id(a)}\nid(b): {id(b)}")
b = 3
print(f"**** after b = 3 ****\nid(a): {id(a)}\nid(b): {id(b)}")
b += 1
print(f"**** after b += 1 ****\nid(a): {id(a)}\nid(b): {id(b)}")
```

## **Mutable**

In Python, ‘mutable’ is the ability of objects to change their values. These are often the objects that store a collection of data.

คำแปล: ค่าของตัวแปร Immutable สามารถเปลี่ยนได้โดยไม่จำเป็นต้องอยู่ทางซ้ายของเครื่องหมาย `=`

```python
a = [2]
print(f"id(a): {id(a)}\t value: {a}")
a[0] = 3
print(f"**** after a[0] = 3 ****")
print(f"id(a): {id(a)}\t value: {a}")
b = a
print(f"**** after b = a ****")
print(f"id(a): {id(a)}\t value: {a}\nid(b): {id(b)}\t value: {b}")
b[0] = 4
print(f"**** after b[0] = 4 ****")
print(f"id(a): {id(a)}\t value: {a}\nid(b): {id(b)}\t value: {b}")
b.append('test')
print(f"**** after b.append('test') ****")
print(f"id(a): {id(a)}\t value: {a}\nid(b): {id(b)}\t value: {b}")
a.remove(4)
print(f"**** after a.remove(4) ****")
print(f"id(a): {id(a)}\t value: {a}\nid(b): {id(b)}\t value: {b}")
```
