# **while**

## **เปรียบเทียบกับ if**

ถ้าตรงตามเงื่อนไขของ if จะทำตามแค่ครั้งเดียว






```python
# if เงิื่อนไข:
#     ทำถ้าเงิือนไขเป็นจริง
# เช่น
# if ฉันหิว:
#     ฉันก็กินข้าว
a = 2
b = 4
print("*** before condition ***")
print("a:", a)
print("b:", b)
if a < b:
     # a is smaller and want to be greater
     a = a + 1
print("*** after condition ***")
print("a:", a)
print("b:", b)
```

## **แต่ถ้าเป็น while**

ถ้าตรงตามเงื่อนไขของ while จะทำไปเรื่อยๆจนกว่าเงื่อนไขจะไม่เป็นจริง






```python
a = 2
b = 4
print("*** before condition ***")
print("a:", a)
print("b:", b)
while a < b:
     # a is smaller and want to be greater
     a = a + 1
print("*** after condition ***")
print("a:", a)
print("b:", b)
```

```python
# ตัวอย่างการใช้ while หาค่าผลรวมของจำนวนใน list
my_list = [4, 2, 4, 1]
i = 0
total = 0
while i < len(my_list):
    print("This round is going to add", total, "with", my_list[i])
    total += my_list[i]
    print("Now total are", total)
    i += 1
print("After while loop, total are", total)
```

# **continue/break**

## **continue จะใช้หยุดการทำงานของ loop รอบนั้นแล้วกลับไปเริ่มต้นทำงานที่ต้น loop ใหม่**

```python
for i in range(10):
    # if we want to exclude any even number
    if i%2 == 0:
        continue
    # if we want to exclude any multiple of 3
    if i%3 == 0:
        continue
    print(i)
```

## **break จะหยุดการทำงานของ for/while แล้วออกไปท้าย loop เลย**

```python
number_count = 0
for i in range(10):
    # if we want to exclude any even number
    if i%2 == 0:
        continue
    # if we want to exclude any multiple of 3
    if i%3 == 0:
        continue
    print(i)
    number_count += 1
    # if we want to print only two numbers
    if number_count == 2:
        break
print("number_count:", number_count)
```

# **ตัวอย่างโจทย์**

เรารู้อยู่กันอยู่แล้วว่า ถ้าเอา a + a/(2^1) + a/(2^2) + a/(2^3) + .... สุดท้ายจะเข้าใกล้ 2a มากขึ้นเรื่อยๆ

คำถามคือมันต้องบวกถึง 2 ยกกำลังเท่าไหร่ ค่าผลบวกของมันถึงจะต่างกับค่า 2 เท่าในระดับที่พอรับได้ (ในที่นี้ขอเป็น 10^-6)

## **ใช้ for แก้ปัญหา**

```python
a = 1
total = 0
for i in range(30):
    total += a/(2**i)
    print("with i:", i, " the total value are", total)
    if 2*a - total < 10e-6:
        break
print("The answer is", i)
```

## **ใช้ while แก้ปัญหา**

```python
a = 1
i = -1
total = 0
while 2*a - total > 10e-6:
    i += 1
    total += a/(2**i)
    print("with i:", i, " the total value are", total)
print("The answer is", i)
```
