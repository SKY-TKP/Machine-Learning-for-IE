# **Tuple**

Tuple จะทำงานเหมือน List ทุกอย่าง ยกเว้นการแก้ไขค่าของสมาชิก

## **การสร้าง Tuple vs การสร้าง List**

```python
l1 = [1, 3, 4]; l2 = [9]
t1 = (1, 3, 4); t2 = (9,) # ถ้าใส่แค่วงเล็บเฉยๆ มันจะกลายเป็นตัวเลขธรรมดา
print(type(l1))
print(type(l2))
print(type(t1))
print(type(t2))
```

## **การใช้งานทั่วไปของ Tuple เทียบกับ List**

```python
t = (11,22,33)
print(len(t))        # 3
print(t[0],t[-1])    # 11 33
print(t[:2])         # (11,22)
print(33 in t)       # True
print(t.index(33))   # 2

# แก้ไขไม่ได้ แต่สร้างใหม่ได้
print(id(t))
#t[0] = 99            # ผิด
t = (99,) + t[1:]    # (99,22,33)
print(t)
print(id(t))
```

## **Nested Tuple**

เราสามารถสร้าง Tuple ซ้อนกันได้

```python
grades = (("6131001021", "A"), ("6130020221", "B"), ("6130150721", "A"))

print(grades[1][1])
grades[1][1] = "C"
print(grades[1][1])
```

ข้อควรระวัง ถึงแม้เราจะไม่สามารถแก้ไขค่าใน Tuple แต่ถ้าเราเอา data structure ตัวอื่นเป็นสมาชิกของ Tuple เรายังสามารถแก้ไขค่าในสมาชิกของ data structure ดังกล่าวได้

```python
grades = (["6131001021", "A"], ["6130020221", "B"], ["6130150721", "A"])

print(grades[1][1])
grades[1][1] = "C"
print(grades[1][1])
```

```python
t = (1, 3, 4, {"A", "D", "Apple"})
t[3].add(23)
t[3] = ["A", "D", "Apple"]
print(t)
```

## **การคืนค่าหลายๆค่าจาก function จะคืนมาในรูปของ Tuple**

```python
def find_min_max(data):
    return min(data), max(data)

x = find_min_max({4, 3, 1, 7})
print(x)
print(type(x))
```

## **Why Tuple over List**

เพราะ Tuple ป้องกันการแก้ไข ฉะนั้นจึงเหมาะแก่การส่งค่าเข้าไปใน function ที่เราไม่ต้องการให้มีการแก้ไขค่า (ตัวอย่างจากการบ้านปี 2021/2)

```python
board = [["", "", "A"], ["", "A", "T"], ["", "", ""]]
```

```python
board = (("", "", "A"), ("", "A", "T"), ("", "", ""))
```

```python
def play2(board):
    board[2][2] = "E"
play2(board)
print(board)
```

Tuple จะมี Bytecode ที่เล็กกว่า (from [Stack Overflow](https://stackoverflow.com/a/22140115))

เพราะ compiler รู้ว่าค่าใน Tuple ไม่เปลี่ยนแน่ๆ มันจึงสามารถหดการทำงานได้

```python
# Bytecode
from dis import dis
dis(compile("(10, 'abc')", '', 'eval'))
dis(compile("[10, 'abc']", '', 'eval'))
```

## **Set Methods**

ตัวแปรประเภทเซตจะมีความสามารถที่เหมือนที่เราเรียนเรื่องเซตมาตอนมัธยม

```python
A = {1,2,3,4,5}
B = {3,4,5,6,7}
```

### **Union**

union.png

```python
C = A.union(B)        # {1,2,3,4,5,6,7}
print(C)
C = A | B             # {1,2,3,4,5,6,7}
print(C)
```

### **Intersection**

intersection.png

```python
C = A.intersection(B) # {3,4,5}
print(C)
C = A & B             # {3,4,5}
print(C)
```

### **Difference**

difference.png

```python
C = A.difference(B)   # {1,2}
print(C)
C = A - B             # {1,2}
print(C)
```

### **Subset**

```python
print(A <= B)
print(A.issubset(B))
```

### **&=, |=, -= (ไม่ออกสอบมั้ง ไปเขียนแบบข้างบนก็ได้)**

```python
A = {1,2,3,4,5}
B = {3,4,5,6,7}

C = A | B             # {1,2,3,4,5,6,7}
print(C)
C = A.copy()
C |= B
print(C)

C = A & B             # {3,4,5}
print(C)
C = A.copy()
C &= B
print(C)

C = A - B             # {1,2}
print(C)
C = A.copy()
C -= B
print(C)
```

## **ความเร็วในการค้นข้อมูลในเซท**

```python
n = 5000
L = []
for i in range(n):
    L.append(i)
S = set()
for i in range(n):
    S.add(i)
```

```python
import time
def search_all(X):
    b = time.time()
    n = len(X)
    for i in range(n):
        if i in X:     # True
            pass
    for i in range(n):
        if (n+1) in X: # False
            pass
    print(time.time() - b)
```

```python
search_all(L)
search_all(S)
```

# **Whatever Comprehension** (ไม่ออกสอบ แต่น่าสนใจ)



บทที่แล้วเรารู้จัก List Comprehension ไปแล้ว คราวนี้มาดูซิว่าถ้าเราเอา Comprehension มาใช้กับ Tuple Set Dict จะเกิดอะไรขึ้น

```python
l = ['oh', 'one', 'two', 'three',
     'four', 'five', 'six', 'seven',
     'eight', 'nine', 'thousand', 'million']
```

```python
# Dict Comprehension
d = {e: len(e) for e in l}
print(d)
print(type(d))
```

```python
# Set Comprehension
s = {e for e in l}
print(s)
print(type(s))
```

```python
# Tuple Comprehension
t = (e for e in l)
print(t)
print(type(t))
for e in t:
    print(e)
t[1] = "abcd"
print(t)
```
