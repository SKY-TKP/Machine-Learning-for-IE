## **การเขียนฟังก์ชัน พารามิเตอร์ และตัวแปร local**

### **ส่วนประกอบในการเขียนฟังก์ชัน**

```
def find_group_index(e, groups):
    for i in range(len(groups)):
        if e in groups[i]:
            return i
    return -1
```

*   def
*   ชื่อฟังก์ชัน
  - กฎการตั้งชือเหมือนการตั้งชื่อตัวแปร  
*   วงเล็บ
*   พารามิเตอร์ (optional)
*   ชุดคำสั่ง
*   return (optional)
  - แปลว่า *คืน* จะคืนสิ่งที่ต่อท้าย return
  - แปลว่า *ย้อนกลับไป*  จะย้อนกลับไปทำคำสั่งต่อจากตำแหน่งที่ฟังก์ชันถูกเรียก
  - ไม่ว่าจะมี return หรือไม่ พอหมดชุดคำสั่งก็จะย้อนกลับไปตำแหน่งที่ฟังก์ชันถูกเรียกใช้เหมือนกัน

## การ **return การไม่คืนผล การคืนผลหนึ่งค่า การคืนผลหลายค่า**
ฟังก์ชันสามารถคืนผลได้หลายรูปแบบตามแต่การเรียกใช้งาน return

````python
def func1():
    # ------------------
    # ประมวลผลอะไรซักอย่าง
    # ------------------
    pass

a = func1()
print("return from func1:", a)
````

````python
def func2():
    # ------------------
    # ประมวลผลอะไรซักอย่าง
    # ------------------
    return

a = func2()
print("return from func2:", a)
````

````python
def func5():
    # ------------------
    # ประมวลผลอะไรซักอย่าง
    # ------------------
    return "I love python", 101, [9, 8, 7]

def func6():
    # ------------------
    # ประมวลผลอะไรซักอย่าง
    # ------------------
    return ("I love python", 101, [9, 8, 7])

a = func5()
print("return from func5:", a)
print("return from func5:", a[2])
a, b, c = func5()
print("return from func5:", b)
a = func6()
print("return from func6:", a)
````

## **ผลจากการเปลี่ยนค่าพารามิเตอร์**
ถ้าแก้ไขค่าพารามิเตอร์ด้วยเครื่องหมาย = ตรงๆ ค่าของตัวแปรนอกฟังก์ชันจะไม่เปลี่ยน

ลองเอา code ข้างล่างไปทดสอบใน [pythontutor](https://pythontutor.com/visualize.html#mode=edit) ดู

````python
def f1(x):
    x = 0
    print("in f1:", x)

a = 9
f1(a)
print("after f1:", a)
````

````python
def f2(x):
    print("in f2 (b4):", x)
    x = [0]*len(x)
    print("in f2 (after):", x)

b = [1, 2, 3, 4]
f2(b)
print("after f2:", b)
````

````python
def f3(x):
    print("in f3:", x)
    for i in range(len(x)):
        x[i] = 0

b = [1, 2, 3, 4]
f3(b)
print("after f3:", b)
````
