### **Local variable**

ตัวแปรภายในฟังก์ขันใดจะเป็นของฟังก์ชันนั้น จะเรียกใช้ตัวแปรในฟังก์ชันอื่นไม่ได้

เพราะตัวแปรในฟังก์ชันนึง จะชื่อซ้ำกับตัวแปรในฟังก์ชันอื่นได้

```python
import math
def sphere_volume(r):
    sv = (4/3) * math.pi * r**3
    return sv

def sphere_surface_area(r):
    sa = 4 * math.pi * r**2
    print("in sphere_surface_area, sa=", sa)
    print("in sphere_surface_area, trying to print x, which is from outside function, x=", x)
    print("in sphere_surface_area, sv=", sv)
    return sa

def main():
    x = 1
    print(sphere_volume(x))
    print(sphere_surface_area(x))

main()
print(sphere_volume(2))
print(sphere_surface_area(2))
```

**การคืนค่าหลายค่า**

```python
def next3integers(x):
    return x+1, x+2, x+3

print(next3integers(4))

y = next3integers(10)
print(type(y), "y:", y)
j, k, l = next3integers(10)
print(type(j), "j:", j)
print(type(k), "k:", k)
print(type(l), "l:", l)
```
