## **lower()/upper()/strip()**



<ตัวแปรสตริง>.lower() จะเป็นการสร้างสตริงใหม่โดยที่ตัวอักษรภาษาอังกฤษเดิมจะกลายเป็นตัวพิมพ์เล็กทั้งหมด

<ตัวแปรสตริง>.upper() จะเป็นการสร้างสตริงใหม่โดยที่ตัวอักษรภาษาอังกฤษเดิมจะกลายเป็นตัวพิมพ์ใหญ่ทั้งหมด

<ตัวแปรสตริง>.strip() จะเป็นการสร้างสตริงใหม่ โดยที่ by default สตริงใหม่จะไม่มี [ASCII whitespace](https://en.wikipedia.org/wiki/Whitespace_character) ทั้งที่นำหน้าและตามหลังสตริง

```python
# lower, upper, strip
s = " Hello World "
print("s:         **" + s + "**")
print("s.lower(): **" + s.lower() + "**")
print("s:         **" + s + "**")
print("s.upper(): **" + s.upper() + "**")
print("s:         **" + s + "**")
print("s.strip(): **" + s.strip() + "**")
print("s:         **" + s + "**")
print("s.strip().lower(): **" + s.strip().lower() + "**")
print("s:         **" + s + "**")
print("s.lower().strip(): **" + s.lower().strip() + "**")
print("s:         **" + s + "**")
```

## **find(e)/find(e,k)**

<ตัวแปรสตริง>.find(e) จะเป็นการค้นหาสตริง e ในตัวแปรสตริง โดยจะ return ค่าเป็น index ที่เจอสตริง e เป็นครั้งแรกหากหาเจอ หากหาไม่เจอจะคืน -1

<ตัวแปรสตริง>.find(e, k) จะเหมือน find(e) แต่ต่างกันตรงที่จะเริ่มการค้นหาที่ตำแหน่ง k

โดยการค้นหาจะเป็น case sensitive



```python
# find(e), find(e, k)
#    0123456789012345678901
s = "AAAbcabcABCAbcabCaBcAb"
print('s.find("b"):   ', s.find("b"))
print('s.find("b", 3):', s.find("b", 3))
print('s.find("b", 4):', s.find("b", 4))
print("***********************")
print('s.find("Abc"):    ', s.find("Abc"))
print('s.find("Abc", 2): ', s.find("Abc", 2))
print('s.find("Abc", 4): ', s.find("Abc", 4))
print('s.find("Abc", 12):', s.find("Abc", 12))
```

# **การอ่านและเขียน file**

การอ่านและเขียน file โดยทั่วไปจะเริ่มด้วยคำสั่ง [open()](https://docs.python.org/3/library/functions.html#open) และปิดด้วย method close()

arguments ของคำสั่ง open()  (มีหลายตัวใน python docs) ที่เราจะสนใจมี 3 ตัว `open(filename, mode='r', encoding=None)`
- filename (ต้องมี) เป็นสตริงระบุว่าเราจะอ่านหรือเขียน file อะไร ซึ่งสามารถจะระบุได้ทั้ง [absolute หรือ relative path](https://www.educative.io/answers/absolute-vs-relative-path)
- mode เป็นสตริงระบุว่าจะเปิด file ไว้ทำอะไรซึ่งในวิชานี้จะพูดถึงแค่ 2 โหมด คือ `"r"` (อ่านอย่างเดียว) กับ `"w"` (เขียนอย่างเดียว) ซึ่งหากไม่ระบุ argument นี้ file จะถูกเปิดด้วยโหมด `"r"`
- encoding (อันนี้ไม่น่าจะออกสอบแต่น่าจะได้เห็นในหลายๆที่) เป็นสตริงระบุว่าจะอ่านหรือเขียน file อย่างไรในกรณีที่ไม่ใช่ภาษาอังกฤษ ซึ่งถ้าจะให้ใช้งานกับภาษาไทยให้ระบุ `encoding="utf-8"`




```python
# ตัวอย่างโค้ดเพื่อเขียน file (เอาไปรันใน Thonny หรืออะไรซักอย่างที่ไม่ใช่ Colab)
f = open("C:\\Users\\jessa\\Documents\\tmp\\test1.txt", "w")
f.write("hello1\n")
f.write("hello2\n")
f.write("hello3\n")
f.close()
# เพื่อความปลอดภัยต้องเรียก method .close() ทุกครั้งเพื่อป้องกันบั๊กอันไม่พึงประสงค์
# เช่น file ไม่ได้รับการเขียน หรือ file โดย lock
```

<font color=red>ข้อควรระวัง</font> การเปิดไฟล์ด้วยโหมด `w` จะลบเนื้อหาในไฟล์เดิมทิ้งหมด ดังนั้นอย่าเปิดไฟล์ที่ต้องการจะอ่านด้วยโหมดนี้

```python
# ตัวอย่างโค้ดเพื่ออ่าน file (เอาไปรันใน Thonny หรืออะไรซักอย่างที่ไม่ใช่ Colab)
# absolute path
f = open("C:\\Users\\jessa\\Documents\\tmp\\test2.txt", "r")
for line in f:
    print(line)
f.close()
# เพื่อความปลอดภัยต้องเรียก method .close() ทุกครั้งเพื่อป้องกันบั๊กอันไม่พึงประสงค์
# เช่น file ไม่ได้รับการเขียน หรือ file โดย lock
```

ในชีวิตจริงจะไม่ค่อยเห็นคนใช้ method .close() เพราะเค้านิยมใช้ `with` ในการเปิด file ซึ่ง `with` จะจัดการปิด file ให้อัตโนมัติ แต่เราจะไม่สอนใช้ในวิชานี้ เพราะอยากจะเน้นไปที่ความสำคัญของการปิด file

**ตัวอย่างเพิ่มเติม**

```python
import numpy as np
import random

# relative path
fout = open("student_score.txt", "w")
n_stu = 50
rng = np.random.RandomState(106)
scores = rng.normal(57, 3, n_stu)
scores[scores>60] = 60
scores = list(scores.astype(int))
S = []
stu_ids = [f"stu{i}" for i in range(1, n_stu+1)]
random.shuffle(stu_ids)
for idx, stu_id in enumerate(stu_ids):
    fout.write(stu_id + "\t" + str(scores[idx]) + "\n")
fout.close()
```
