# **เปรียบเทียบ dict กับ list**






โดยทั่วไปจะทำงานคล้ายๆ list แต่จะมีส่วนที่ต่างกันอยู่บ้าง

## **list ว่าง vs dict ว่าง**

```python
# list ว่าง
x = []
print("empty list:", x)
```

```python
# dict ว่าง
x = {}
print("empty dict:", x)
```

## **การสร้าง list vs dict**

```python
# การสร้าง list ที่มีสมาชิกขึ้นมาเลย
l = ["A", "B", "C"]
print(l)
```

```python
# การสร้าง dict ที่มีสมาชิกขึ้นมาเลย
d = {"A": 10, "B": 20, "C": 30} # dict จะเก็บข้อมูลแบบ key-value pair (โปรดสังเกตว่าวงเล็บจะเป็น {})
print(d)
```

## **การอ่านค่าสมาชิกและการกำหนดค่า**

ใน list วิธีการอ่านค่าหรือกำหนดค่าจะทำผ่าน list[index]

```python
l = ["A", "B", 987]
print("value at index 1:", l[1])
print("************")
print("value in l[2] before reassigned: ", l[2])
l[2] = "D"
print("value in l[2] after reassigned: ", l[2])
#print(l[3])
```

ใน dict วิธีการอ่านค่าหรือกำหนดค่าจะทำผ่าน dict[key]

```python
d = {"A": 10, "B": 20, "C": 30}
print('value at key "A":', d["A"])
print("************")
print("value in d before reassigned: ", d)
d["B"] = "A CompProg แน่นอน"
print("value in d after reassigned", d)
#print(d["a"]) # จะมี error คล้ายๆ list ถ้าพยายามจะเข้าถึงตำแหน่งที่ไม่มี key นั้นอยู่
```

key ที่เราเรียนกันจะเป็นสตริงหรือตัวเลข ตามใจเรา

```python
d = {100: 10, "B": 20, "6430000121": "grade C", 0.1: "12"}
print(d)
print(d["6430000121"])
print(d[0.1])
#print(d["grade C"])
```

ส่วน value ก็เป็นได้ทุกอย่างที่สมาชิกของ list จะเป็นได้

```python
d = {100: 10, "B": 20, "6430000121": ["male", "165 cm", "A CompProg", "F Physics"]}
print(d["6430000121"])
print(type(d["6430000121"]))
```

แน่นอนว่า value ก็เป็น dict ได้เช่นกัน

```python
d = {100: 10, "B": 20, "6430000121": {"gender": "male", "height": 165, "2110101": "A", "2304103": "D"}}
print(d["6430000121"])
print(type(d["6430000121"]))
```

## **การเพิ่มค่า**

เราใช้ list.append() สำหรับการเพิ่มค่าเข้าไปใน list ที่มีอยู่

```python
# หรือสร้างด้วยการเพิ่มสมาชิกเข้าไปจาก list เดิม
l = []
l.append(10)
l.append("I love Python")
l.append("A")
print(l)
```

สำหรับ dict เราสามารถเพิ่มค่าในวิธีการเดียวกับการเปลี่ยนแปลงค่า ถ้า key นั้นมีอยู่แล้วจะเปลี่ยนแปลงค่า ถ้า key นั้นไม่เคยมีอยู่ จะเป็นการเพิ่มสมาชิกเข้าไปใน dict

```python
d = {}
d[100] = 10
d["B"] = 20
d["6430000121"] = "grade C"
d[0.1] = "12"
print(d)
d["6430000221"] = "grade B+" # ถ้า key ยังไม่เคยมีใน dict มาก่อนจะเป็นการเพิ่มค่า แต่ถ้ามีอยู่แล้วก็จะเป็นการแก้ไขค่า
print("after adding 6430000221")
print(d)
```

## **การค้นหาค่า**

สำหรับ list เราใช้ in เพื่อตรวจสอบว่าค่าที่เราต้องการค้นหาเป็นสมาชิกของ list หรือไม่ และใช้ `.index` ในการค้นว่าค่าที่เราต้องการหาอยู่ที่ index ไหน

```python
l = ["Eren", "Armin", "Historia"]
print("Armin" in l)
print(l.index("Armin"))
print("Annie" in l)
#print(l.index("Annie"))
```

การตรวจสอบว่า `key` ที่ต้องการค้นหามีอยู่ dict หรือไม่ จะเหมือนกับ list คือใช้คำสั่ง in เหมือนกัน แต่เราไม่มีคำสั่งในการค้น `value`

แต่การหา index จะแตกต่างคือ เราจะเข้าถึง `key` ได้เลยโดยไม่ต้องใช้คำสั่งในการค้นหา แต่สำหรับ `value` ก็ต้องใช้ loop ในการค้นหา เพราะไม่มีคำสั่งรองรับ

```python
d = {100: 10, "B": 20, "6430000121": {"gender": "male", "height": 165, "2110101": "A", "2304103": "F"}}
print(100 in d)
print(10 in d)
print("6430000121" in d)
```

## **Loop**

```python
d = {100: 10, "B": 20, "6430000121": {"gender": "male", "height": 165, "2110101": "A", "2304103": "F"}}
for key in d:
    print(key)
```

# **For in keys()/values()/items()**

## **keys()**

```python
d = {100: 10, "B": 20, "6430000121": {"gender": "male", "height": 165, "2110101": "A", "2304103": "F"}}
for key in d.keys():
    print(key)
```

## **values()**

```python
d = {100: 10, "B": 20, "6430000121": {"gender": "male", "height": 165, "2110101": "A", "2304103": "F"}}
for value in d.values():
    print(value)
```

## **items()**

```python
d = {100: 10, "B": 20, "6430000121": {"gender": "male", "height": 165, "2110101": "A", "2304103": "F"}}
for key, value in d.items():
    print(key, value)
```

# **ความเร็ว**

```python
# สร้างลิสต์ l ที่มีสมาชิก "เยอะๆ"
l = []
for i in range(20000):
    l.append('{:09d}'.format(i))
```

```python
# สร้าง dict d ที่มีสมาชิก "เยอะๆ"
d = {}
for i in range(20000):
    val = '{:09d}'.format(i)
    d[val] = val
```

```python
print(len(l))
print(len(d))
```

```python
import time
def search_all(X):
    b = time.time()
    n = len(X)
    for i in range(n):
        if i in X:     # True
            pass
    for i in range(n):
        if (n+1) in X: # False
            pass
    print(time.time() - b)
```

```python
search_all(l)
```

```python
search_all(d)
```
