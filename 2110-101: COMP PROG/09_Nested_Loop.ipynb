# **สรุปการใช้ Nested Control Structures**






## **Nested While**

```python
a = 0
while a < 5:
    print("a:", a)
    a += 1
```

```python
b = 3
while b > 0:
    print(" b:", b)
    b -= 1
```

```python
a = 0
while a < 5:
    b = 3
    while b > 0:
        print("a:", a, " b:", b)
        b -= 1
    a += 1
```

## **Nested For**

```python
for alphabet in "ABCD":
    for i in range(3):
        print(alphabet, i)
```

## **Break from nested loop**

สามารถทำได้ด้วยทั้งการใช้ break และ function

```python
# การใช้ break สามารถใช้ตัวแปรช่วยการเก็บสถานะในการ break
c = 0
enough = False
for alphabet in "ABCD":
    for i in range(5):
        print(alphabet, i)
        c += 1
        if c >= 13:
            enough = True
            break
    if enough:
        break
```

```python
# ถ้าใช้ function จะใช้ return ออกจาก function แทนการ break
def count_until_enough():
    c = 0
    for alphabet in "ABCD":
        for i in range(5):
            print(alphabet, i)
            c += 1
            if c >= 13:
                return
count_until_enough()
```

## **การสร้าง nested list**

```python
# จะรับ input เข้ามาเรื่อยๆ แต่ละบรรทัดจะเป็นชื่อและปีเกิดคั่นด้วยช่องว่าง ซึ่งจะหยุดก็ต่อเมื่อ input เป็น q
# สุดท้ายจะแสดงผลเรียงตามปีเกิดและชื่อ
result = []
x = input()
while x != "q":
    name, year = x.split()
    year = int(year)
    result.append([year, name])
    x = input()
print(result)
result.sort()
print(result)
for year, name in result:
    print(name, year)
```

## **Nested list as matrix**

```python
def print_matrix( M ):
    if len(M) == 1:
        print(M)
    else:
        print("[" + str(M[0]))
        for i in range(1,len(M)-1):
            print(" " + str(M[i]))
        print(" " + str(M[-1]) + "]")
M = [[1,2,3,4],[2,2,1,3],[2,6,7,7]]
print_matrix(M)
def sum_matrix_by_column(M):
    result = []
    for i in range(len(M[0])):
        total = 0
        for j in range(len(M)):
            total += M[j][i]
        result.append(total)
    return result
sum_matrix = sum_matrix_by_column(M)
print(sum_matrix)
```

## **List Comprehension**

เป็นการเขียนคำสั่งสร้างลิสต์ที่ **สั้น** และมี **ประสิทธิภาพ**



```python
# แปลงทุกค่าใน x เป็นอีกอย่าง
x = [2, -1, 3]
t1 = []
for e in x:
    t1.append(2*e)
print(t1)
# ------- แบบ list comprehension ---------
t2 = [2*e for e in x]
print(t2)
```

```python
# เลือกบางค่าในลิสต์ x
x = [2, -1, 3]
t1 = []
for e in x:
    if e >= 0:
        t1.append(e)
print(t1)
# ------- แบบ list comprehension ---------
t2 = [e for e in x if e >= 0]
print(t2)
```

```python
# ทั้งแปลงทั้งเลือก
x = [2, -1, 3]
t1 = []
for e in x:
    if e >= 0:
        t1.append(2*e)
print(t1)
# ------- แบบ list comprehension ---------
t2 = [e*2 for e in x if e >= 0]
print(t2)
```

## **ความเร็วของ list comprehension**

```python
x = []
for i in range(1000000):
    x.append(i)
```

```python
%%timeit
l = []
for e in x:
    l.append(e)
```

```python
%%timeit
l = [e for e in x]
```

```python
import timeit
def for_loop(n):
    t = []
    for i in range(n):
        t.append(n)
    return t
def comprehension(n):
    return [n for i in range(n)]
def time(func):
    print(timeit.timeit(func+"(1000000)",
          globals=globals(), number=100))
time("for_loop")      # 9.2609192
time("comprehension") # 4.7720880999999995
```
